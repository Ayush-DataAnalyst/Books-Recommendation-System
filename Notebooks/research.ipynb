```python
ratings_books = ratings.merge(books, on='ISBN')
```

```python
number_ratings = ratings_books.groupby('Title')['Book-Rating'].count().reset_index()
```

```python
number_ratings.head()
```

```python
number_ratings.rename(columns={'Book-Rating':'num_of_ratings'},inplace=True)
```

```python
number_ratings.head()
```

```python
final_ratings = ratings_books.merge(number_ratings, on='Title')
```

```python
final_ratings.head()
```

```python
book_counts = final_ratings['Title'].value_counts()
popular_books = book_counts[book_counts >= 50].index
filtered_books = final_ratings[final_ratings['Title'].isin(popular_books)]
```

```python
user_counts = final_ratings['User-ID'].value_counts()
active_users = user_counts[user_counts >= 10].index
filtered_data = filtered_books[filtered_books['User-ID'].isin(active_users)]

book_pivot = filtered_data.pivot_table(index='Title', columns='User-ID', values='Book-Rating')
book_pivot
```

```python
final_ratings.head()
```

```python
final_ratings.rename(columns=lambda x: x.strip(), inplace=True)
final_ratings.dropna(subset=['Author', 'Publisher', 'Iamage-URL'], inplace=True)
final_ratings.drop_duplicates(inplace=True)
final_ratings.isnull().sum()
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

# Testing Model

```python
def recommend_book(books_name):
    book_id = np.where(book_pivot.index == books_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)

    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == books_name:
                print(f"You Searched '{books_name}'\n")
                print("The suggestion Books Are: \n")
            else:
                print(j)
```

```python
books_name = "2nd Chance"
recommend_book(books_name)
```
